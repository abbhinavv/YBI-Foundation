```python
df_features =df[[ 'Movie_Genre','Movie_Keywords','Movie_Tagline','Movie_Cast','Movie_Director']].fillna('')
```

```python
df_features.shape
```

```python
df_features
```

```python
X = df_features['Movie_Genre']+'  '+df_features['Movie_Keywords']+' '+df_features['Movie_Tagline']+' '+df_features['Movie_Cast']+' '+df_features['Movie_Director']
```

```python
X.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
```

```python
tfdf =TfidfVectorizer()
```

```python
X=tfdf.fit_transform(X)
```

```python
X.shape
```

```python
print(X)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(X)
```

```python
similarity_score.shape
```

```python
Favorite_Movie_Name =input('Enter your Favorite movie name :')
```

```python
All_movies_title_List =df['Movie_Title'].tolist()
```

```python
import difflib
```

```python
Movie_Recommendation =difflib.get_close_matches(Favorite_Movie_Name,All_movies_title_List)
print(Movie_Recommendation)
```

```python
Close_Match =Movie_Recommendation[0]
print(Close_Match)
```

```python
index_of_Match_movie =df[df.Movie_Title==Close_Match]['Movie_ID'].values[0]
print(index_of_Match_movie)
```

```python
Recommendation_Score = list(enumerate(similarity_score[index_of_Match_movie ]))
```

```python
print(Recommendation_Score )
```

```python
len(Recommendation_Score )
```

```python
Sorted_Similar_Movies =sorted(Recommendation_Score,key=lambda x:x[1],reverse=True)
print(Sorted_Similar_Movies)
```

```python
print('Top 11 Movies Suggested for You :\n')
i=1
for Movie in Sorted_Similar_Movies:
  index = Movie[0]
  title_from_index = df[df.index==index]['Movie_Title'].values[0]
  if(i<11):
    print(i,'.',title_from_index)
    i+=1
```
